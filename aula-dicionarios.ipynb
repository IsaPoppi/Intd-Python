#Dicionários

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
```

#Métodos aplicados a Dicionário

copy() -> Returns a copy of the dictionary

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.copy()

print(x)
```

fromkeys() ->	Returns a dictionary with the specified keys and value

```python
#criando um dicionário com 3 chaves e valores iguais a 0
x = ('key1', 'key2', 'key3')
print(type(x))
y = 0

thisdict = dict.fromkeys(x, y)

print(thisdict)
```

get()	-> Returns the value of the specified key

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.get("model")

print(x)
```

items() ->	Returns a list containing a tuple for each key value pair

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.items()

#print(x)

for i,j in x:
  print(j)
```

keys()-> Returns a list containing the dictionary's keys

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.keys()

print(x)
```

pop()	-> Removes the element with the specified key

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.pop("model")

print(car)
```

popitem()	-> Removes the last inserted key-value pair

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.popitem()

print(car)
```

update()	-> Updates the dictionary with the specified key-value pairs

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.update({"color": "White"})

print(car)
```

values()	-> Returns a list of all the values in the dictionary

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.values()

print(x)
```

clear() -> Removes all the elements from the dictionary

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.clear()

print(car)
```
