# biblioteca BeautifulSoup
## instalando a biblioteca e capturando o primeiro html

```python
# importando as bibliotecas
from bs4 import BeautifulSoup
import requests
```

```python
site= "http://www.pudim.com.br/"
```

```python
html=requests.get(site)
```

```python
print(html.text)
```

```python
# Coleta os dados do site
html=requests.get(site).content
html
```

```python
# Formatando os dados
dados=BeautifulSoup(html, 'html.parser')
print(dados.prettify())
```

##Coletando elementos HTML específicos

```python
# Busca por tag HTML
busca_por_tag=dados.find("img")
busca_por_tag
```

```python
# Busca por id
busca_por_id=dados.find(id="ABC")
```

```python
# Busca por class
busca_por_class=dados.find(class_="email")
busca_por_class
```

```python
# Busca por class com html oganizado
print(busca_por_class.prettify())
```

```python
# Buscando todos os elementos
busca_tudo=dados.find_all("div")
busca_tudo
```

## Coletando apenas os textos

```python
# coletando a class email
email=dados.find("div",class_="email")
email
```

```python
# coletando a tag dentro da div email
link=email.find("a")
print(link)
```

```python
# coletando apenas o texto
print(link.text)
```

```python
# coletando o endereço do link
print("3",link.attrs['href'])
```
